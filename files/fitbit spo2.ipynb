```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file
file_path = 'SpO2.xlsx'
df = pd.read_excel(file_path)

# Extract relevant columns
x = pd.to_datetime(df['timestamp'])
y = df['average_value']

# Calculate mean and standard deviation
mean_y = y.mean()
std_y = y.std()

# Calculate the lines for 2 standard deviations above and below
y_plus_2std = mean_y + 2 * std_y
y_minus_2std = mean_y - 2 * std_y

# Create scatter plot
plt.scatter(x, y, label='Data Points')

# Add line charts for 2 standard deviations above and below
plt.plot(x, [y_plus_2std] * len(x), color='r', linestyle='--', label='+2 Std Dev')
plt.plot(x, [y_minus_2std] * len(x), color='g', linestyle='--', label='-2 Std Dev')

# Customize plot
plt.title('Scatter Plot with 2 Standard Deviation Lines')
plt.xlabel('Timestamp')
plt.ylabel('Average Value')
plt.legend()
plt.xticks([])  # Remove x-axis labels
plt.tight_layout()

# Show plot
plt.show()
```
