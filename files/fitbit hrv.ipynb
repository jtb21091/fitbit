```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'hrv.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
df.head()

# Function to plot scatter with avg and std lines as line graphs
def plot_with_line_graphs(data, column_name):
    plt.figure(figsize=(10, 6))
    plt.scatter(data.index, data[column_name], label=f'{column_name} Data', alpha=0.5)
    
    # Calculate mean and std deviation
    mean_value = data[column_name].mean()
    std_dev = data[column_name].std()
    plus_2std = mean_value + 2 * std_dev
    minus_2std = mean_value - 2 * std_dev

    # Plot mean and std lines
    plt.plot(data.index, [mean_value]*len(data), color='green', linestyle='-', label=f'Mean ({mean_value:.2f})')
    plt.plot(data.index, [plus_2std]*len(data), color='red', linestyle='--', label=f'Mean + 2 STD ({plus_2std:.2f})')
    plt.plot(data.index, [minus_2std]*len(data), color='blue', linestyle='--', label=f'Mean - 2 STD ({minus_2std:.2f})')

    plt.title(f'Scatter Plot with Mean and 2 STD Lines for {column_name}')
    plt.xlabel('Index')
    plt.ylabel(column_name)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot RMSSD with line graphs
plot_with_line_graphs(df, 'rmssd')

# Plot NREMHR with line graphs
plot_with_line_graphs(df, 'nremhr')
```
